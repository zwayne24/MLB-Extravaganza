```python
# beautiful soup https://www.espn.com/nba/standings

url = 'https://www.basketballforcoaches.com/basketball-quotes/'
headers = {
    'User-Agent': 'Mozilla/5.0'
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# get every basketball quote
paragraphs = []
for p in soup.find_all('p'):
    paragraphs.append(p.text)
```

```python
# escape " " in quotes
paragraphs = [p.replace('“', '') for p in paragraphs]
paragraphs = [p.replace('”', '') for p in paragraphs]
```

```python
# find only quotes that contain a dash
quotes = []
for p in paragraphs:
    if '–' in p:
        quotes.append(p)
```

```python
quotes_df = pd.DataFrame(columns=['Quotes', 'Speaker'])
for q in quotes:
    if len(q.split('–')) > 2:
        pass
    else:
        quotes_df = quotes_df.append({'Quotes': q.split('–')[0], 'Speaker': q.split('–')[1]}, ignore_index=True)
```

```python
# remove leading and trailing white spaces
quotes_df['Quotes'] = quotes_df['Quotes'].str.strip()
quotes_df['Speaker'] = quotes_df['Speaker'].str.strip()
```

```python
quotes_df.to_csv('quotesMaster.csv', index=False)
```

```python
all_nba_2_first = all_nba_2[all_nba_2.columns[all_nba_2.columns.get_level_values(0) == 'First team']]
all_nba_2_first.columns = all_nba_2_first.columns.droplevel(0)
all_nba_2_first['Team Level'] = 'First Team'
all_nba_2_first['Season'] = all_nba_2[('Season', 'Season')]
all_nba_2_first['Positions'] = all_nba_2[('Positions', 'Positions')]

all_nba_2_second = all_nba_2[all_nba_2.columns[all_nba_2.columns.get_level_values(0) == 'Second team']]
all_nba_2_second.columns = all_nba_2_second.columns.droplevel(0)
all_nba_2_second['Team Level'] = 'Second Team'
all_nba_2_second['Season'] = all_nba_2[('Season', 'Season')]
all_nba_2_second['Positions'] = all_nba_2[('Positions', 'Positions')]

all_nba_3_first = all_nba_3[all_nba_3.columns[all_nba_3.columns.get_level_values(0) == 'First team']]
all_nba_3_first.columns = all_nba_3_first.columns.droplevel(0)
all_nba_3_first['Team Level'] = 'First Team'
all_nba_3_first['Season'] = all_nba_3[('Season', 'Season')]
all_nba_3_first['Positions'] = all_nba_3[('Positions', 'Positions')]

all_nba_3_second = all_nba_3[all_nba_3.columns[all_nba_3.columns.get_level_values(0) == 'Second team']]
all_nba_3_second.columns = all_nba_3_second.columns.droplevel(0)
all_nba_3_second['Team Level'] = 'Second Team'
all_nba_3_second['Season'] = all_nba_3[('Season', 'Season')]
all_nba_3_second['Positions'] = all_nba_3[('Positions', 'Positions')]

all_nba_3_third = all_nba_3[all_nba_3.columns[all_nba_3.columns.get_level_values(0) == 'Third team']]
all_nba_3_third.columns = all_nba_3_third.columns.droplevel(0)
all_nba_3_third['Team Level'] = 'Third Team'
all_nba_3_third['Season'] = all_nba_3[('Season', 'Season')]
all_nba_3_third['Positions'] = all_nba_3[('Positions', 'Positions')]

all_nba_4_first = all_nba_4[all_nba_4.columns[all_nba_4.columns.get_level_values(0) == 'First team']]
all_nba_4_first.columns = all_nba_4_first.columns.droplevel(0)
all_nba_4_first['Team Level'] = 'First Team'
all_nba_4_first['Season'] = all_nba_4[('Season', 'Season')]
all_nba_4_first['Positions'] = ["TBD"] * len(all_nba_4_first)

all_nba_4_second = all_nba_4[all_nba_4.columns[all_nba_4.columns.get_level_values(0) == 'Second team']]
all_nba_4_second.columns = all_nba_4_second.columns.droplevel(0)
all_nba_4_second['Team Level'] = 'Second Team'
all_nba_4_second['Season'] = all_nba_4[('Season', 'Season')]
all_nba_4_second['Positions'] = ["TBD"] * len(all_nba_4_second)


all_nba_4_third = all_nba_4[all_nba_4.columns[all_nba_4.columns.get_level_values(0) == 'Third team']]
all_nba_4_third.columns = all_nba_4_third.columns.droplevel(0)
all_nba_4_third['Team Level'] = 'Third Team'
all_nba_4_third['Season'] = all_nba_4[('Season', 'Season')]
all_nba_4_third['Positions'] = ["TBD"] * len(all_nba_4_third)

all_nba = pd.concat([all_nba_1_first, all_nba_1_second, all_nba_2_first, all_nba_2_second, all_nba_3_first, all_nba_3_second, all_nba_3_third, all_nba_4_first, all_nba_4_second, all_nba_4_third])
```

```python
# replace †, §, ^, * with empty string
all_nba['Players'] = all_nba['Players'].str.replace('†', '').str.replace('§', '').str.replace('^', '').str.replace('*', '')
# remove parentheses and numbers
all_nba['Players'] = all_nba['Players'].str.replace(r"\(.*\d+.*\)", "")
# remove any leading or trailing white spaces
all_nba['Players'] = all_nba['Players'].str.strip()
```

```python
all_nba = all_nba.drop_duplicates(subset=['Players', 'Season'])
```

```python
# for each player, get Player Name, Team List, Times on First Team, Times on Second Team, Times on Third Team, first season, last season
result = all_nba.groupby('Players').agg(
    Team_List=('Teams', lambda x: list(x.unique())),
    Times_First_Team=('Team Level', lambda x: (x == 'First Team').sum()),
    Times_Second_Team=('Team Level', lambda x: (x == 'Second Team').sum()),
    Times_Third_Team=('Team Level', lambda x: (x == 'Third Team').sum()),
    First_Season=('Season', 'min'),
    Last_Season=('Season', 'max'),
    Positions=('Positions', lambda x: list(x.unique()))
).reset_index()
```

```python
# remove TBD from position list
result['Positions'] = result['Positions'].apply(lambda x: [i for i in x if i != 'TBD'])
```

```python
# write to csv
result.to_csv('allNBA.csv', index=False)
```

```python
import cv2
import mediapipe as mp
import numpy as np

df = pd.read_excel("player_schedule.xlsx")

for index, row in df.iterrows():
    # Load image
    image = cv2.imread(row['image_path'])
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Init Face Mesh
    mp_face_mesh = mp.solutions.face_mesh
    face_mesh = mp_face_mesh.FaceMesh(static_image_mode=True)

    # Process image
    results = face_mesh.process(image_rgb)

    if results.multi_face_landmarks:
        height, width, _ = image.shape
        mask = np.zeros((height, width), dtype=np.uint8)

        for face_landmarks in results.multi_face_landmarks:
            # Select landmark indexes covering the face/skin area
            face_idxs = list(range(0, 468))  # you can narrow this to facial contour indexes
            points = []
            for idx in face_idxs:
                x = int(face_landmarks.landmark[idx].x * width)
                y = int(face_landmarks.landmark[idx].y * height)
                points.append((x, y))

            # Convex hull of the facial points
            points = np.array(points, dtype=np.int32)
            hull = cv2.convexHull(points)
            cv2.fillConvexPoly(mask, hull, 255)

        # Black out skin area on original image
        skin_silhouette = image.copy()
        skin_silhouette[mask == 255] = (0, 0, 0)

        cv2.imwrite(row['player_name'] + '_silhouette.jpg', skin_silhouette)
    else:
        print("No face detected.")
```

```python
import os
import requests
import pandas as pd
from bs4 import BeautifulSoup
from pybaseball import playerid_lookup

# Load player list from Excel
df = pd.read_excel("player_list.xlsx")  # must have columns: first_name, last_name

# Make output folder
os.makedirs("images", exist_ok=True)

for _, row in df.iterrows():
    first = row['first_name']
    last = row['last_name']
    print(f"Processing: {first} {last}")

    try:
        # Lookup player ID
        data = playerid_lookup(last, first)
        if data.empty:
            print(f"❌ Could not find ID for {first} {last}")
            continue

        player_id = data.key_bbref.iloc[0]
        first_letter = player_id[0]
        url = f"https://www.baseball-reference.com/players/{first_letter}/{player_id}.shtml"

        # Get player page
        response = requests.get(url)
        if response.status_code != 200:
            print(f"❌ Failed to fetch {url}")
            continue

        soup = BeautifulSoup(response.text, 'html.parser')
        media_div = soup.find('div', class_='media-item multiple')

        if media_div:
            img_tag = media_div.find('img')
            if img_tag and 'src' in img_tag.attrs:
                img_url = img_tag['src']
                img_data = requests.get(img_url).content

                filename = f"images/{first} {last}.jpg"
                with open(filename, 'wb') as f:
                    f.write(img_data)

                print(f"✅ Saved: {filename}")
            else:
                print(f"❌ No image tag found for {first} {last}")
        else:
            print(f"❌ No media section found for {first} {last}")

    except Exception as e:
        print(f"⚠️ Error processing {first} {last}: {e}")
```
